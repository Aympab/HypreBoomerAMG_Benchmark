```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image, set_matplotlib_formats 
# set_matplotlib_formats('pdf') # toggle vector graphics for a crisp plot!
set_matplotlib_formats('svg')
```

```python
agglevel = pd.read_csv("./strong/aggressive_level/agglevel_np28_nx8096.csv", header=0, dtype=float, names=['paramvalue','time','ite','resi'])

plt.plot(agglevel['paramvalue'][:-5], np.log(agglevel['time'][:-5]), 'x-')
plt.ylabel("$log-Duration$")
plt.xlabel("$Aggressive\ level$")
plt.xticks(range(1,10))
plt.grid()
```

```python
cortype = pd.read_csv("./strong/coarsen_type/ctype_np28_nx8096.csv", header=0, dtype=float, names=['paramvalue','time','ite','resi'])

print(cortype)
plt.plot(cortype['paramvalue'], np.log(cortype['time']), 'x-')
plt.ylabel("$log-Duration$")
plt.xlabel("$Coarsen\ Type$")
plt.xticks(cortype['paramvalue'], ['CLJP', 'RugeS', 'RugeS2', 'Falg.', 'rCLJP', 'PMIS', 'PMIS', '', 'opRugeS'])
plt.grid()
```

```python
relaxtype = pd.read_csv("./strong/relax_type/rtype_np28_nx8096.csv", header=0, dtype=float, names=['paramvalue','time','ite','resi'])

print(relaxtype)
plt.plot(relaxtype['paramvalue'], np.log(relaxtype['time']), 'x-')
plt.ylabel("$log-Duration$")
plt.xlabel("$Smoother\ type$")
plt.xticks(relaxtype['paramvalue'], ['Jacobi', 'GS', 'SORfor', 'SORback', 'chaotic GS', 'SSOR'])
plt.grid()
```

# Best Param

```python
bp = pd.read_csv("./strong/params/best_params_nx8096.csv", header=0, dtype=float, names=['np','time','ite','resi'])
dp = pd.read_csv("./strong/params/default_params_nx8096.csv", header=0, dtype=float, names=['np','time','ite','resi'])

print(bp)
print(dp)

plt.plot(bp['np'], np.log(bp['time']), 'bx-', label="Fitted param")
plt.plot(dp['np'], np.log(dp['time']), 'gx-', label="Default param")

plt.ylabel("$log-Duration$")
plt.xlabel("$Nb proc$")
plt.xticks(bp['np'], ['1', '2', '4', '6', '8', '12', '16', '20', '24', '28'])
plt.legend()
plt.grid()
```
